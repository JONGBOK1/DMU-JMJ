```python
import folium

map = folium.Map(location=[37.58, 126.89],zoom_start=11,
                 zoom_control=False,control_scale=True)

mk = folium.Marker([37.6507, 126.7789], tooltip='더클라임 일산점',max_width=100,
                    icon=folium.Icon(icon='star'))
mk.add_to(map)
mk2 = folium.Marker([37.5608, 126.8338], tooltip='더클라임 마곡점',
                     icon=folium.Icon(color='orange', icon='cloud'))
mk2.add_to(map)
mk3 = folium.Marker([37.5716, 126.9998], tooltip='클라이밍파크 종로점',
                     icon=folium.Icon(color='green',icon='heart'))
mk3.add_to(map)
mk4 = folium.Marker([37.485, 126.8966], tooltip='서울숲클라이밍 구로점',
                     icon=folium.Icon(color='green',icon='heart'))
mk4.add_to(map)
mk5 = folium.Marker([37.5546, 126.922], tooltip='볼더프렌즈 클라이밍',
                     icon=folium.Icon(icon='star'))
mk5.add_to(map)
mk6 = folium.Marker([37.5607, 126.9431], tooltip='신촌담장',
                     icon=folium.Icon(icon='star'))
mk6.add_to(map)
mk7 = folium.Marker([37.5174, 126.8996], tooltip='서울숲클라이밍 영등포점',
                     icon=folium.Icon(icon='flag'))
mk7.add_to(map)
mk8 = folium.Marker([37.5699, 126.9901], tooltip='서울숲클라이밍 종로점',
                     icon=folium.Icon(color='red',icon='book'))
mk8.add_to(map)
mk9 = folium.Marker([37.5547, 126.9203], tooltip='더클라임 홍대점',
                     icon=folium.Icon(color='orange',icon='book'))
mk9.add_to(map)
cc1 = folium.Circle([37.4836, 126.9016], tooltip='닭장맥주 구로디지털단지점',
                     color='green',radius=250,fill=True)
cc1.add_to(map)
cc2 = folium.Circle([37.5497, 126.9207], tooltip='카츠미 홍대점',
                     color='green',radius=150,fill=True)
cc2.add_to(map)
cc3 = folium.Circle([37.5712, 126.0002], tooltip='계동치킨 종로5가점',
                     color='green',radius=300,fill=True)
cc3.add_to(map)
cc4 = folium.Circle([37.5833, 126.9996], tooltip='유캔하이',
                     color='red',radius=200)
cc4.add_to(map)

from folium.plugins import MarkerCluster
mc=MarkerCluster().add_to(map)

folium.Marker([37.5547, 126.9203]).add_to(mc)
folium.Marker([37.5497, 126.9207]).add_to(mc)
folium.Marker([37.5546, 126.922]).add_to(mc)
folium.Marker([37.5716, 126.9998]).add_to(mc)
folium.Marker([37.5699, 126.9901]).add_to(mc)
folium.Marker([37.5712, 126.0002]).add_to(mc)
folium.Marker([37.5833, 126.9996]).add_to(mc)
folium.Marker([37.5174, 126.8996]).add_to(mc)
folium.Marker([37.485, 126.8966]).add_to(mc)
folium.Marker([37.5607, 126.9431]).add_to(mc)
map.save('MapEx_20242530.html')
map
```
